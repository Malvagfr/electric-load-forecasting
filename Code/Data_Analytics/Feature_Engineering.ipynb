```python
# Import libraries 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import dataframe 
df_electricity_demand=pd.read_csv("../../Data/Intermediate_Data/electricity_demand.csv")
df_electricity_demand
```

## Creating calendar features

```python
df_electricity_demand['Time']=pd.to_datetime(df_electricity_demand['Time'], format="%Y-%m-%d %H:%M:%S")
```

```python
df_electricity_demand["Week_day"]=df_electricity_demand.Time.dt.day_name()
```

```python
#sabado parcial, domingo total, nacional total, bank holiday 
df_electricity_demand['Week_day_category']=np.where(df_electricity_demand["Time"].dt.dayofweek>4, 
    'Weekend','Week')
```

```python
df_electricity_demand["Bank_Holiday_Weight"]=df_electricity_demand["Country_Bank_Holiday"]\
+df_electricity_demand["Partial_Bank_Holiday_Weight"]
```
