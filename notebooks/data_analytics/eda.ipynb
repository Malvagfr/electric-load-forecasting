## Libraries import

```python
# Ignore warnings
import warnings
warnings.filterwarnings('ignore')

# Libraries Imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Functions definition

```python
# Read csv
def read_csv(path):
    return pd.read_csv(path)
```

```python
# Drop columns
def drop_columns(df,columns):
    return df.drop(columns, 1) 
```

## Analysis

```python
# Import data 
df_electricity_demand=read_csv("../../data/intermediate_data/electricity_demand.csv")
df_electricity_demand=drop_columns(df_electricity_demand,'Unnamed: 0')
```

### 4. Features correlation

```python
# Create a correlation matrix
corr_matrix = df_electricity_demand.corr()

# Plot the correlation using seaborn
plt.subplots(figsize = (10, 8))
sns.heatmap(corr_matrix, annot = True)
plt.show()
```

Most importants features according to correlation with the demand:
